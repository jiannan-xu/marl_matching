```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
x = np.arange(-20,20,0.001)

plt.plot(x,norm.pdf(x,loc=1,scale=2),label='loc=8,scale=1')
plt.plot(x,norm.pdf(x,loc=5,scale=2),label='loc=9,scale=2')
plt.plot(x,norm.pdf(x,loc=10,scale=2),label='loc=10,scale=2')

```
